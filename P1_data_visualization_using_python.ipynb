```python
import pandas as pd
import matplotlib.pyplot as plt

#load csv data
file_path="C:/Users/hibas/Desktop/Self_Learning/Project_1_Data_visualization_using_python/courses.csv"
data=pd.read_csv(file_path)

#plot a bar chart
plt.figure(figsize=(8,6))
plt.bar(data['Course'],data['Students'],color='red')

#add labels and title
plt.xlabel('Courses')
plt.ylabel('Number of Students')
plt.title('Students Enrolled as per Courses')

#show the plot
plt.show()
```
